```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# 1. Entorno de Bandit
class Bandit:
    def __init__(self, k=10):
        self.k = k
        self.probs = np.random.rand(k)  # probabilidades verdaderas de cada brazo

    def pull(self, arm):
        # 2. Tirar del brazo: recompensa 1 con probabilidad específica, sino 0
        return 1 if np.random.rand() < self.probs[arm] else 0
```

```python

# 3. Agente épsilon-greedy
class Agent:
    def __init__(self, k, epsilon):
        self.k = k
        self.epsilon = epsilon
        self.estimates = np.zeros(k)  # 4. Recompensas estimadas iniciales
        self.counts = np.zeros(k)     # 4. Contador de extracciones

    def select_arm(self):
        # 5. Selección de brazo
        if np.random.rand() < self.epsilon:
            return np.random.randint(self.k)
        return np.argmax(self.estimates)

    def update_estimates(self, arm, reward):
        # 6. Actualizar estimación incremental
        self.counts[arm] += 1
        n = self.counts[arm]
        self.estimates[arm] += (reward - self.estimates[arm]) / n
```

```python
# 7. Inicializar
np.random.seed(42)
bandit = Bandit(k=10)
agent = Agent(k=10, epsilon=0.1)
```

```python
# 8. Simulación de 1,000 iteraciones
n_iters = 1000
cumulative_rewards = np.zeros(n_iters)
total_reward = 0

```

```python
for i in range(n_iters):
    arm = agent.select_arm()                  # 9a. Seleccionar brazo
    reward = bandit.pull(arm)                 # 9b. Tirar del brazo
    agent.update_estimates(arm, reward)       # 9c. Actualizar estimaciones
    total_reward += reward
    cumulative_rewards[i] = total_reward      # 10. Seguimiento de recompensa acumulada

print(f"Recompensa acumulada total tras {n_iters} iteraciones: {total_reward} con epsilon={agent.epsilon}")


```

```python
# 11. Graficar recompensa acumulada
plt.figure()
plt.plot(np.arange(1, n_iters + 1), cumulative_rewards)
plt.xlabel('Iteración')
plt.ylabel('Recompensa acumulada')
plt.title('Recompensa acumulada a lo largo del tiempo (epsilon=0.1)')
plt.show()

```

```python
# 12. Graficar estimaciones vs verdaderas probabilidades
plt.figure()
x = np.arange(len(agent.estimates))
width = 0.35
plt.bar(x - width/2, agent.estimates, width, label='Estimado')
plt.bar(x + width/2, bandit.probs, width, label='Verdadero')
plt.xlabel('Brazos')
plt.ylabel('Valor')
plt.title('Estimaciones vs Probabilidades Verdaderas (epsilon=0.1)')
plt.legend()
plt.show()
```
